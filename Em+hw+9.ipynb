```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
B = 2

T_I = []
R_I = []
X = []
x = 0
dx = np.pi/1000
while x < np.pi/100:
    A = (1 - ((1/2)*np.sin(np.rad2deg(x)))**2)**(1/2)/(np.cos(np.rad2deg(x)))
    R = (1-(A*B))/(1+(A*B))
    R_I.append(R)
    U = 2/(1+(A*B))
    T_I.append(U)
    X.append(x)
    x += dx
plt.plot(X, R_I, label="ER/EI")
plt.plot(X, T_I, label="ET/EI")
plt.legend()

    
```
